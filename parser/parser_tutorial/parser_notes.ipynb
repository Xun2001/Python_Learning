## parser_1.py

```python
import argparse
parser = argparse.ArgumentParser()
parser.add_argument("echo")
args = parser.parse_args() #调用 parse_args() 方法解析命令行参数，并将解析结果存储在 args 对象中。
print(args.echo)
```

### 位置参数

```python
parser.add_argument("echo") #这里是位置参数
```

位置参数（Positional Arguments）  

定义：位置参数是必须提供的参数，它们按照在命令行中的位置顺序进行解析。

特点：
- 不需要前缀标志（如 - 或 --）。
- 必须提供，否则会抛出错误。
- 按照定义的顺序解析。

如上所述的 echo 就是位置参数，只要是第一个位置输入的参数，都会被解析为  
parser中的echo参数，之后想调用该参数的时候，需要指定他的名字为 echo；  
  
等于是 定义了一个echo 变量，按位置来输入参数对该变量进行赋值； 

**Note:**

- 位置参数按位置解析
- 但是位置参数和可选参数之间不做约束，可选参数不在意位置，且可以在位置参数前或后

### 选择参数

顾名思义，该参数可以不赋值使用，但是位置参数必须指定

#### 可选项 -- True/False

```python
parser.add_argument('-gpu','--use_gpu',help='whether to use gpu',
                    default=True,type=bool)
```

```python
parser.add_argument("-v","--verbose", help="increase output verbosity",
                    action="store_true")
```
